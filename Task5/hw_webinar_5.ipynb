Ответьте на вопросы:
- В чем принципиальное отличие гибридных рекомендательных систем от коллаборативной филтьтрации?
- Приведите 2-3 примера задач, в которых необходимо использовать гибридные системы

1. Гибридная система может включать в себя коллаборативную фильтрацию и сожержать еще какую-либо фильтрацию например content based. Основной проблемой коллаборативных систем является холодный старт. Гибридная же система т.к. содержит еще какой-либо механизм фильтрации и способна решать проблему холодного старта.

2. Задачи с холодным стартом. Например, что рекомендовать новому пользователю без истории покупок. Или, например появилось очень много совершенно новой продукции (например новогодней) без истории продаж и ее нужно как то рекомендовать. Например Netflix рекомендует фильмы выделяя похожих пользователей в кластеры и рекомендуя каждому кластеру одни и те же фильмы, а если фильм или пользователь новый, то рекомендует фильмы с высоким рейтингом.
